```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
df=sklearn.datasets.load_breast_cancer()
```

```python
# converting into dataframe
data=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
# adding target columns into dataset
data['label']=df.target
```

```python
# Now splitting the dataframe

X = data.drop(columns='label',axis=1)
Y=data['label']
```

```python
X
```

```python
Y
```

```python
# Now split the data into train and test

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print('x_train : ',x_train.shape)
print('x_test: ',x_test.shape)
print("X: ", X.shape)
```

```python
# Now preprocess the data
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()

New_X_train=scale.fit_transform(x_train)
New_X_test=scale.fit_transform(x_test)
```

```python
# importing tesnorflow and keras

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Now setting the layers in the Neural Network , Making ANN

model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(20,activation='tanh'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
# Now compile the Neuarl Network

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Training the Neural Network

history=model.fit(New_X_train,y_train,validation_split=0.2,epochs=50)
```

```python
# Visualizing the data

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training_data', 'Validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# Now accuracy on the test data

model.evaluate(New_X_test,y_test)
```
